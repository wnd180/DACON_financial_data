```python
import pandas as pd
from pandas import DataFrame
from pandas import Series
import numpy as np
import matplotlib.pyplot as plt

pd.options.display.float_format = '{:.5f}'.format
pd.options.display.float_format = '{:,.0f}'.format

df_basic = pd.read_excel("1-1-1. 세입 수입 예산편성현황(총액).xlsx")
df_mod = df_basic[['회계연도','소관명','국회확정금액(천원)']]

df_mod
```

```python
grouped_sum = df_mod['국회확정금액(천원)'].groupby([df_mod['회계연도'],df_mod['소관명']]).sum()
DataFrame(grouped_sum)
```

```python
result = grouped_sum.unstack()
pd.set_option('display.max_columns',None)
result
```
